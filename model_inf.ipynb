```python
import pandas as pd
import numpy as np
import pickle
import json
```

```python
# Load model
with open("model.pkl", "rb") as file_1:
    model = pickle.load(file_1)

# Load scaler
with open("scaler.pkl", "rb") as file_2:
    scaler = pickle.load(file_2)
# Load encoder
with open("encoder.pkl", "rb") as file_3:
    encoder_n = pickle.load(file_3)
# Load numeric list

with open("num_col.txt", "rb") as file_4:
    num_col = json.load(file_4)
# load categoric list
with open("cat_col.txt", "rb") as file_5:
    cat_col = json.load(file_5)
```

```python
data_inf = { 
 'city' : 'Tangerang',
 'property_type' : "rumah" ,
 'bedrooms' : 3 ,
 'bathrooms' : 2 ,
 'land_area' : 181 ,
 'building_area' : 182, 
 'floors' : 2 ,
 'maid_bedrooms' : 1 ,
 'maid_bathrooms' : 1 ,
 'certificate' : "shm - sertifikat hak milik ",
 'voltage' : 5500 ,
 'building_age' : 1 ,
 'year' : 2021 ,
 'condition' : "baru", 
 'garage' : 1,
  'carport' :1 }
# ubah ke dataframe
data_inf =pd.DataFrame([data_inf])

#cek
data_inf
```

```python
data_inf_num = data_inf[num_col]
data_inf_cat = data_inf[cat_col] 
display(data_inf_num)
display(data_inf_cat)
```

```python
data_inf_num_scaled = scaler.transform(data_inf_num)

data_inf_cat_encoded = encoder_n.transform(data_inf_cat).toarray()
```

```python
data_inf_final = np.concatenate([data_inf_num_scaled, data_inf_cat_encoded],axis=1)
data_inf_final
```

```python
y_pred_inf = model.predict(data_inf_final)

y_pred_inf
```
